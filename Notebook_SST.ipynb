# Modelo Autoencoder

Preprocesamiento de las imágenes

```python
from datetime import datetime

from os import listdir
lista_img = listdir('/content/sst_dataset')

def ordenar_lista_cronologicamente(lista_img):
    return sorted(lista_img, key=lambda x: datetime.strptime(x.split('.')[1][:6], "%Y%m"))


import cv2
import numpy as np
import pandas as pd
from datetime import datetime


def generar_dataset_completo(lista_sst, size=(16, 42)):
    lista_sst = ordenar_lista_cronologicamente(lista_sst)
    fechas_sst = [f.split('.')[1][:6] for f in lista_sst]

    dataset = []
    registros = []

    for sst_img in lista_sst:
        img_sst = cv2.imread('sst_dataset/' + sst_img)


        img_sst = cv2.cvtColor(img_sst, cv2.COLOR_BGR2GRAY)
        img_sst = img_sst.astype(np.float32) / 255.0
        img_sst = img_sst[..., np.newaxis]

        dataset.append(img_sst)
        registros.append({
            'fecha': sst_img.split('.')[1][:6],
            'nombre_sst': sst_img
        })

    dataset = np.array(dataset)
    df_nom_fechas = pd.DataFrame(registros)

    return dataset, df_nom_fechas
```

Bloque Encoder

```python
from tensorflow.keras import layers, models
from tensorflow.keras import backend as K

def encoder(input_shape = (16,42,1), latent_dim = 10):
  inputs = layers.Input(shape=input_shape)
  #Primera capa
  x = layers.Conv2D(32, 3, strides=2, padding='same', activation='relu')(inputs)  #(8, 21, 32)
  x = layers.BatchNormalization()(x)


  encoder_output_shape = K.int_shape(x)[1:]
  #Aplanar para preparar la salida del encoder
  x = layers.Flatten()(x)
  encoder_output = layers.Dense(latent_dim)(x)

  modelo_encoder = models.Model(inputs=inputs, outputs=encoder_output, name="Encoder")
  return encoder_output_shape, modelo_encoder
```

Bloque Decoder

```python
def decoder(encoder_output_shape, latent_dim = 10):
  inputs = layers.Input(shape= (latent_dim,))

  #Primera capa lineal
  x = layers.Dense(np.prod(encoder_output_shape))(inputs)
  x = layers.Reshape(encoder_output_shape)(x)

  #Segunda capa
  x = layers.Conv2DTranspose(32, 3, strides=2, padding='same', activation='relu')(x)  #(16, 42, 32)
  x = layers.BatchNormalization()(x)

  #Hacemos Conv2DTranspose = upsampling + Conv2D, última capa activación sigmoid -> salida entre 0 y 1.
  output_decoder = layers.Conv2DTranspose(1, (3,3), padding='same', activation = 'sigmoid') (x)

  modelo_decoder = models.Model(inputs = inputs, outputs = output_decoder, name = "Decoder")
  return modelo_decoder
```

Bloque Autoencoder: Encoder + Decoder

```python
def autoencoder (input_shape = (16, 42, 1), latent_dim = 10):
  encoder_output_shape, modelo_encoder = encoder(input_shape, latent_dim)
  modelo_decoder = decoder(encoder_output_shape, latent_dim)
  #para que la parte del encoder
  #quede nombrada como Encoder y así podamos extraer los pesos desde el AE
  inputs = layers.Input(shape=input_shape)
  encoder_output = modelo_encoder(inputs)  #uso del encoder como capa-modelo
  decoder_output = modelo_decoder(encoder_output)

  modelo_autoencoder = models.Model(inputs = inputs, outputs = decoder_output, name = "Autoencoder")

  return modelo_encoder, modelo_decoder, modelo_autoencoder
```

Compilación del modelo Autoencoder

```python
from tensorflow.keras.optimizers import Adam
def compilar(modelo, learning_rate = 0.0005, loss = 'mse'):
  optimizer = Adam(learning_rate=learning_rate)
  modelo.compile(optimizer=optimizer, loss=loss)
  return modelo
```

Definir el entrenamiento del modelo Autoencoder

```python
from tensorflow.keras.callbacks import CSVLogger, EarlyStopping, ModelCheckpoint

def entrenar_modelo(modelo, x_train, batch_size =8, epochs = 100, validation_split = 0.1, usar_callbacks = True,
                    checkpoint_ruta = 'sst_archivos/autoencoder.weights.h5', log_ruta = 'sst_archivos/log_entrenamiento.csv'):
  callbacks = []
  if usar_callbacks:
    callbacks = [CSVLogger(log_ruta),
                 ModelCheckpoint(checkpoint_ruta,
                                 monitor = "val_loss",
                                 save_weights_only=True,
                                 save_best_only=True,
                                 verbose=1),
                                EarlyStopping(monitor = "val_loss", patience = 25)]
  #No ponemos steps_per_epoch porque steps_per_epoch = len(x_train) // batch_size esto solo hace falta cuando el
  #dataset es muy grande o los datos provienen de un dataflow.
  historico = modelo.fit(x_train,
                       x_train, #al ser autoencoder el output debe ser el mismo tras todas las transformaciones
                       batch_size=batch_size,
                       epochs=epochs,
                       validation_split=validation_split,
                       callbacks=callbacks)
  return historico
```

Entrenar el modelo Autoencoder

```python
#Ejecutar AE
dataset_compacto, df_info = generar_dataset_completo(lista_img)

modelo_encoder, modelo_decoder, modelo_autoencoder = autoencoder()
modelo_autoencoder = compilar(modelo_autoencoder)
historico = entrenar_modelo (x_train = dataset_compacto,modelo = modelo_autoencoder)

#Cargar el modelo autoencoder con los mejores pesos entrenados
modelo_autoencoder.load_weights('sst_archivos/autoencoder.weights.h5')
#Extraer los pesos de la parte del Encoder
modelo_encoder.set_weights(modelo_autoencoder.get_layer("Encoder").get_weights())
modelo_encoder.save("sst_archivos/encoder_entrenado.h5")

#Cargar el modelo encoder para obtener el espacio latente de las imgs
from tensorflow.keras.models import load_model

encoder = load_model('sst_archivos/encoder_entrenado.h5')
espacio_latente = encoder.predict(dataset_compacto)
```

###### Contrastes de hipótesis

Crear función con la que aplicar contrates de hipótesis de normalidad, homocedasticidad y comparación entre grupos

```python
#!pip install pingouin
```

```python
from scipy.stats import shapiro, levene, f_oneway, mannwhitneyu
import pingouin as pg

def analizar_contraste(df, variable, fase_col='fases'):
    print(f"\nVariable: {variable}")

    normales = []
    for fase in df[fase_col].unique():
        grupo = df[df[fase_col] == fase][variable]
        stat, p = shapiro(grupo)
        normales.append(p > 0.05)
        print(f" - Normalidad en {fase}: p-valor = {p:.4f}")

    normalidad = all(normales)

    if normalidad:
        grupo1 = df[df[fase_col] == df[fase_col].unique()[0]][variable]
        grupo2 = df[df[fase_col] == df[fase_col].unique()[1]][variable]
        stat, p_levene = levene(grupo1, grupo2)
        print(f" - Homocedasticidad (Levene): p-valor = {p_levene:.4f}")

        if p_levene > 0.05:
            stat, p_anova = f_oneway(grupo1, grupo2)
            print(f"ANOVA clásico: p-valor = {p_anova:.4f}")
        else:
            resultado = pg.welch_anova(dv=variable, between=fase_col, data=df)
            print(f"Welch: p-valor = {resultado['p-unc'].values[0]:.4f}")

    else:
        grupo1 = df[df[fase_col] == df[fase_col].unique()[0]][variable]
        grupo2 = df[df[fase_col] == df[fase_col].unique()[1]][variable]
        stat, p_mw = mannwhitneyu(grupo1, grupo2, alternative='two-sided')
        print(f"Mann-Whitney U: p-valor = {p_mw:.4f}")
```
